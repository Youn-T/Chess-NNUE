```python
def evaluate(board): # Placeholder for a more complex evaluation function
    return random.uniform(-1,1)
```

```python
def minimax(board, depth, alpha=float('-inf'), beta=float('inf')):
    # depth -= 1
    
    if depth == 0 or board.is_game_over():
        result = evaluate(board)
        # print("result:", result)
        return result
    
    moves = list(board.legal_moves)
    moves_weight = [float('-inf') if board.turn == chess.WHITE else float('inf')] * len(moves)
    for i, move in enumerate(moves):
        board.push(move)
        
        minimax_value = minimax(board, depth -1, alpha, beta)
        board.pop()
        
        if board.turn == chess.WHITE:
            alpha = max(alpha, minimax_value)
        else:            
            beta = min(beta, minimax_value)
        if beta <= alpha:
            break
        moves_weight[i] = minimax_value
        
    print(f"Depth: {depth}, Result: {max(moves_weight) if board.turn == chess.WHITE else min(moves_weight)}, Weights: {moves_weight} Turn {'White' if board.turn == chess.WHITE else 'Black'}")
    return max(moves_weight) if board.turn == chess.WHITE else min(moves_weight)
```

```python
def board_to_input(board):
    input_vector = np.zeros(64*12)
    for square in chess.SQUARES:
        piece = board.piece_at(square)
        if piece is not None:
            piece_type = piece.piece_type - 1
            color_offset = 0 if piece.color == chess.WHITE else 6
            input_vector[square * 12 + color_offset + piece_type] = 1
    return input_vector
```

### Imports

```python
import numpy as np
```

### Constants

```python
NEURONE_COUNT = 512
INPUT_COUNT = 768
BATCH_SIZE = 4096
LEARNING_RATE = 0.01
```

### Key Functions

```python
def ReLU(x):
    return np.maximum(0, x)
    
def CReLU(x):
    return np.clip(x, 0, 1)

def ReLU_derivative(x):
    return np.where(x > 0, 1, 0)

def CReLU_derivative(x):
    return np.where(x > 0 and x < 1, 1, 0)

def Sigmoid(x):
    return 1 / (1 + np.exp(-x))

def Sigmoid_derivative(x):
    return Sigmoid(x) * (1 - Sigmoid(x))


def MSE(prediction, target):
    return np.square( prediction - target )
```

### Weights

```python
W1 = np.random.uniform(-1,1,(INPUT_COUNT, NEURONE_COUNT))
b1 = np.random.uniform(-1,1,(1, NEURONE_COUNT))
W2 = np.random.uniform(-1,1,(NEURONE_COUNT, 1))
b2 = np.random.uniform(-1,1,(1,1))

def init_weights(input_size, hidden_size, output_size):
    W1 = np.random.uniform(-1,1,(input_size, hidden_size))
    b1 = np.random.uniform(-1,1,(1, hidden_size))
    W2 = np.random.uniform(-1,1,(hidden_size, output_size))
    b2 = np.random.uniform(-1,1,(1, output_size))
    return W1, b1, W2, b2
```

### Training

#### Forward Pass

```python
def forward_pass(input, W1, b1, W2, b2):
    
    # HIDDEN LAYER
    Z1 = np.dot(input,W1) + b1
    A1 = ReLU(Z1)

    # OUTPUT LAYER
    Z2 = np.dot(A1 , W2) + b2
    A2 = Sigmoid(Z2)
    return A1, Z1, A2

# A1, Z1, A2 = forward_pass(np.zeros((1, INPUT_COUNT)),W1, b1, W2, b2)    
# A2
```

#### Backward Pass

```python
def backward_pass(input, W1, b1, A1,Z1, W2, b2, A2, labels, alpha):
    d2 = 2 * ( A2 - labels ) * Sigmoid_derivative(A2)    
    dW2 = np.dot(A1.T, d2)
    db2 = np.sum(d2, axis=1, keepdims=True)

    d1 = np.dot(d2, W2.T) * ReLU_derivative(Z1)
    dW1 = np.dot(input.T, d1)
    db1 = np.sum(d1, axis=1, keepdims=True)

    b2 = b2 - alpha * db2
    W2 = W2 - alpha * dW2
    b1 = b1 - alpha * db1
    W1 = W1 - alpha * dW1
    return W1, b1, W2, b2

# W1, b1, W2, b2 = backward_pass(np.zeros((1, INPUT_COUNT)), W1, b1, A1,Z1, W2, b2, A2, 0.5, 0.1)
```

#### Launcher

```python
W1, b1, W2, b2 = init_weights(INPUT_COUNT, NEURONE_COUNT, 1)
logging = False

for epoch in range(30):
    A1, Z1, A2 = forward_pass(BOARDS[:BATCH_SIZE],W1, b1, W2, b2)
    
    rounded_predictions = np.round(A2.flatten())
    accuracy = np.mean(rounded_predictions == LABELS[:BATCH_SIZE])
    if logging:
        print(f"Epoch {epoch}: Predictions: {rounded_predictions}, Labels: {LABELS[:BATCH_SIZE]}, Accuracy: {accuracy}")
    else:
        print(f"Epoch {epoch}: Accuracy: {accuracy}")
    W1, b1, W2, b2 = backward_pass(BOARDS[:BATCH_SIZE], W1, b1, A1,Z1, W2, b2, A2, (LABELS[:BATCH_SIZE]).reshape(-1, 1), LEARNING_RATE)
```

```python
EPOCHS = 10

for epoch in range(EPOCHS):
    indices = np.arange(len(BOARDS))
    np.random.shuffle(indices)
    
    BOARDS_SHUFFLED = BOARDS[indices]
    LABELS_SHUFFLED = LABELS[indices]
    
    for i in range(0, len(BOARDS), BATCH_SIZE):
        if (i+BATCH_SIZE) > len(BOARDS):
            break
        X_batch = BOARDS_SHUFFLED[i:i+BATCH_SIZE]
        y_batch = LABELS_SHUFFLED[i:i+BATCH_SIZE].reshape(-1, 1)
        
        A1, Z1, A2 = forward_pass(X_batch, W1, b1, W2, b2)
        
        rounded_predictions = np.round(A2.flatten())
        accuracy = np.mean(rounded_predictions == y_batch.flatten())
        print(f"Epoch {epoch}, Batch {i//BATCH_SIZE}: Accuracy: {accuracy}")
        
        W1, b1, W2, b2 = backward_pass(X_batch, W1, b1, A1,Z1, W2, b2, A2, y_batch, LEARNING_RATE)
        
    print(f"End of Epoch {epoch}")
```
